```python
import pandas as pd
from bs4 import BeautifulSoup 
import requests
import re
```

```python
url='https://www.racing-statistics.com/en/events/2005-australian-grand-prix'
```

```python
html_content=requests.get(url).text
soup=BeautifulSoup(html_content, 'lxml')
```

```python
table = soup.find_all('fieldset', attrs={'class':'block'})[3]
table
```

```python
body = table.find_all("tr")
head = body[0]
body_rows = body[1:]
len(body_rows)

headings = []
for item in head.find_all("th"):
    item = (item.text).rstrip("\n")
    headings.append(item)
if '' in headings:
    headings.remove('')
headings
```

```python
headings
```

```python
all_rows=[]

for row_num in range(len(body_rows)):
    row=[]
    for row_item in body_rows[row_num].find_all("td"):
        row.append(re.sub("(\xa0)|(\n)|,","",row_item.text))
    if row[0]=='':
        row[0]='DNF'
    while '' in row:
        row.remove('')
    all_rows.append(row)
```

```python
headings,all_rows
```

```python
df = pd.DataFrame(data=all_rows,columns=headings)
df
# df.to_csv('Belgian Grand Prix.csv', index=False)
```
